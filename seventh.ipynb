```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
print(X_train)
```

```python
X_train = pd.DataFrame(
    X_train, 
    columns = X_test.columns)

print(X_train)
```

```python
X_train.head()
```

```python
X_test = sc.transform(X_test)
X_test = pd.DataFrame(
    X_test, columns = X_train.columns
)
```

```python
X_train_np = X_train.values
y_train_np = y_train.values
X_test_np = X_test.values
y_test_np = y_test.values
```

```python
print(type(X_train_np))
```

```python
print(X_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
```

```python
model = Sequential()

model.add(
    Dense(
        units = 6, activation = 'relu',
        kernel_initializer = 'uniform',
        input_dim = 11
    )
)
model.add(Dropout(rate = 0.3))
```

```python
model = Sequential()

model.add(
    Dense(
        units = 6, activation = 'relu',
        kernel_initializer = 'uniform',
    )
)
model.add(Dropout(rate = 0.3))
```

```python
model.add(
    Dense(
        units = 12, activation = 'tanh',
        kernel_initializer = 'uniform'
    )
)
```

```python
model.add(
    Dense(
        units = 6, activation = 'tanh',
        kernel_initializer = 'uniform'
    )
)
```

```python
model.add(
    Dense(
        units = 1, activation = 'sigmoid',
        kernel_initializer = 'uniform'
    )
)
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(
    X_train_np, y_train_np, 
    epochs = 400, 
    # 계산할 때 몇개 단위로 계산할 것인가
    batch_size = 20
)
```

```python
y_pred_class = model.predict(X_test_np)
y_pred_prob = model.predict_proba(X_test_np)
```

```python
from sklearn.metrics import roc_curve

# fpr(거짓 양성률): 실제로는 해당사항이 없지만 해당 사항이 있다고 판정을 내리는 경우
fpr, tpr, thresholds = roc_curve(
    y_test, y_pred_prob
)
```

```python
print(fpr)
```

```python
print(tpr)
```

```python
print(thresholds)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr, tpr)
plt.title("ROC Curve")
plt.xlabel("False Positive Rate (1 - Specificity)")
plt.ylabel("True Positive Rate (Specificity)")
plt.grid(True)
plt.show()
```

```python
def optimum_threshold(my_threshold):
    print(
        "Specificity:", 
        tpr[thresholds > my_threshold][-1]
         )
    print(
        "Specificity:", 
        1 - fpr[thresholds > my_threshold][-1]
    )
```

```python
print(tpr[thresholds > 0.5])
```

```python
print(tpr[thresholds > 0.5][-1])
```

```python
optimum_threshold(0.3)
```

```python
optimum_threshold(0.5)
```

```python
optimum_threshold(0.01)
```
